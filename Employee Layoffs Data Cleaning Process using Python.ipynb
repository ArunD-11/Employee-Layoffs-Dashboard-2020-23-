# ***3) Eliminating Missing values in a data using mean mode imputation -***

```python
data.isna().sum()
```

```python
for i in data.columns:
    if(data[i].dtypes == 'object'):
        x = data[i].mode()[0]
        data[i]=data[i].fillna(x)
    else:
        x = data[i].mean()
        data[i]=data[i].fillna(x)
```

```python
data
```

```python
data.isna().sum()
```

# ***4) Converting data into continuous and categorical data -***

```python
cat = []
con = []

for i in data.columns:
    if(data[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
```
